## 딕셔너리_연습문제 1

문제: 다음딕셔너리에대해물음에답하라.

 days = {'January':31, 'February':28, 'March':31, 'April':30,
 'May':31, 'June':30, 'July':31, 'August':31,
 'September':30, 'October':31, 'November':30, 'December':31}

• 사용자가월을입력하면해당월에일수를출력하라
• 알파벳순서로모든월을출력하라

• 일수가31인월을모두출력하라

• 월의일수를기준으로오름차순으로(key-value) 쌍을 출력하라

• 사용자가월을3자리만입력하면월의일수를출력하라.(Jan, Feb 등)

날짜: 2025.9.16

학과: 컴퓨터공학과

학번: 202210558

저자: 박로사

```python
# 초기 딕셔너리
days = {
    'January': 31, 'February': 28, 'March': 31, 'April': 30,
    'May': 31, 'June': 30, 'July': 31, 'August': 31,
    'September': 30, 'October': 31, 'November': 30, 'December': 31
}

# 1) 사용자가 월을 입력하면 해당 월의 일수를 출력
month = input("월 이름 입력 (예: March): ")
if month in days:
    print(f"{month}는 {days[month]}일까지 있습니다.")
else:
    print("해당 월은 없습니다.")

print("-" * 40)

# 2) 알파벳 순서로 모든 월 출력
print("알파벳 순서 월 목록:")
for m in sorted(days.keys()):
    print(m)

print("-" * 40)

# 3) 일수가 31인 월 출력
print("일수가 31일인 월:")
for m, d in days.items():
    if d == 31:
        print(m)

print("-" * 40)

# 4) 월의 일수를 기준으로 오름차순 출력
print("일수 기준 오름차순 정렬:")
for m, d in sorted(days.items(), key=lambda x: x[1]):
    print(f"{m}: {d}")

print("-" * 40)

# 5) 사용자가 3자리만 입력해도 일수를 출력 (예: Jan, Feb)
abbr = input("월 약자 입력 (예: Jan): ")
found = False
for m, d in days.items():
    if m.startswith(abbr):
        print(f"{m}는 {d}일까지 있습니다.")
        found = True
        break

if not found:
    print("해당 약자와 일치하는 월이 없습니다.")
```
